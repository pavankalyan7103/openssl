```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.action_chains import ActionChains
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
import time
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

# Email configuration
def send_alert_email():
    sender_email = "rixdorferstreet@gmail.com"
    receiver_email = "pavankalyan7103@gmail.com"
    password = "Rixdorferpk@123"

    message = MIMEMultipart("alternative")
    message["Subject"] = "Odds Alert"
    message["From"] = sender_email
    message["To"] = receiver_email

    text = "The specified odds are now available on the betting site."
    part1 = MIMEText(text, "plain")
    message.attach(part1)

    with smtplib.SMTP_SSL("smtp.example.com", 465) as server:
        server.login(sender_email, password)
        server.sendmail(sender_email, receiver_email, message.as_string())
        print("Alert email sent!")
```

```python
# Set up Chrome options
chrome_options = Options()
chrome_options.add_argument("--start-maximized")  # Optional: start Chrome maximized

# Initialize WebDriver with Chrome (try one of these methods)

try:
    # Method 1: Using webdriver_manager with Service parameter (recommended method)
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
except Exception as e1:
    print("Method 1 failed:", e1)
    try:
        # Method 2: Using webdriver_manager without specifying `service` (alternative method)
        driver = webdriver.Chrome(ChromeDriverManager().install(), chrome_options=chrome_options)
    except Exception as e2:
        print("Method 2 failed:", e2)
        try:
            # Method 3: Manually specify ChromeDriver service with ChromeDriver path
            driver_path = "/Users/pavankalyan/Downloads/chromedriver-mac-arm64"  # Replace with actual path
            driver = webdriver.Chrome(service=Service(driver_path), options=chrome_options)
        except Exception as e3:
            print("Method 3 failed:", e3)
            raise e3  # Re-raise the last exception if all methods fail



# Go to the login page
driver.get("https://jewelexch.com")

# Fill in login details
username = "h120jwl96"
password = "Spri1111"

try:
    # Locate the username and password fields and enter credentials
    username_field = driver.find_element(By.XPATH, '//*[@id="root"]/div[2]/div/div/div[2]/form/div[1]/input')  # Replace with actual XPath or ID
    password_field = driver.find_element(By.XPATH, '//*[@id="root"]/div[2]/div/div/div[2]/form/div[2]/input')  # Replace with actual XPath or ID
    login_button = driver.find_element(By.XPATH, '//*[@id="root"]/div[2]/div/div/div[2]/form/button')  # Replace with actual XPath or ID

    username_field.send_keys(username)
    password_field.send_keys(password)
    login_button.click()

    # Wait for the page to load after login
    time.sleep(5)  # Adjust based on loading time

    # Close the popup
    try:
        close_popup_button = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/div[1]/button') # Replace with actual XPath of the close button
        close_popup_button.click()
        time.sleep(1)  # Give a moment for the popup to close
    except Exception as e:
        print("Popup not found or already closed:", e)

    # Scroll down to find the Teenpatti 1-Day game
    game_found = False
    while not game_found:
        try:
            # Locate the Teenpatti 1-Day game element
            teenpatti_1day_link = driver.find_element(By.XPATH, '//*[@id="root"]/div[2]/div[1]/div[3]/div/div[3]/div[16]/a/div[1]')  # Replace with actual XPath for Teenpatti 1-Day
            # Scroll to the game element and click it
            ActionChains(driver).move_to_element(teenpatti_1day_link).perform()
            teenpatti_1day_link.click()
            game_found = True
        except:
            # Scroll down by a bit if the game is not yet found
            driver.execute_script("window.scrollBy(0, 300);")
            time.sleep(1)  # Small delay to let the page load more content

    # Wait for the game page to load
    time.sleep(5)

    while True:
        # Locate the odds element (adjust the selector based on the actual HTML structure)
        odds_element = driver.find_element(By.XPATH, '//*[@id="odds_xpath"]')  # Replace with actual XPath
        odds = float(odds_element.text)

        # Check if the odds meet your criteria
        if odds == 1.15:
            send_alert_email()
            break

        # Wait before checking again
        time.sleep(30)  # Check every 30 seconds
except Exception as e:
    print("An error occurred:", e)
finally:
    driver.quit()
```
